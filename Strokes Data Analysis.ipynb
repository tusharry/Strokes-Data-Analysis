```python
data.fillna(np.mean(data['bmi']), inplace = True)
```

```python
data.isnull().sum()
```

```python
numerical_values = data[['age', 'bmi', 'avg_glucose_level']]
categorical_values = data[['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status', 'stroke']]
```

```python
plt.figure(figsize = (8, 6))
sns.kdeplot(data = numerical_values, shade = True)
plt.grid()
plt.title('Strokes according to AGE, BMI, AVG_GLUCOSE_LEVEL')
```

```python
plt.figure(figsize = (8, 6))
sns.kdeplot(x = data.loc[data['stroke'] == 1]['age'], shade = True, label = 'Stroke')
sns.kdeplot(x = data.loc[data['stroke'] == 0]['age'], shade = True, label = 'No Stroke')
plt.legend(loc = 'upper left')
plt.grid()
plt.title('Strokes according to AGE')

plt.figure(figsize = (8, 6))
sns.kdeplot(data['bmi'], shade = True)
sns.kdeplot(x = data.loc[data['stroke'] == 1]['bmi'], shade = True, label = 'Stroke')
sns.kdeplot(x = data.loc[data['stroke'] == 0]['bmi'], shade = True, label = 'No Stroke')
plt.legend(loc = 'best')
plt.grid()
plt.title('Strokes according to BMI')

plt.figure(figsize = (8, 6))
sns.kdeplot(x = data.loc[data['stroke'] ==1]['avg_glucose_level'], shade = True, label = 'Stroke')
sns.kdeplot(x = data.loc[data['stroke'] ==0]['avg_glucose_level'], shade = True, label = 'No Stroke')
plt.legend(loc = 'best')
plt.grid()
plt.title('Strokes according to AVG_GLUCOSE_LEVEL')
```

```python
plt.figure(figsize = (12, 7))
sns.scatterplot(x = data['age'], y = data['bmi'], hue = categorical_values['stroke'], palette="OrRd")
plt.legend(loc = 'best')
plt.grid()
plt.title('BMI -> AGE')

plt.figure(figsize = (12, 7))
sns.scatterplot(x = data['age'], y = data['avg_glucose_level'], hue = categorical_values['stroke'], palette="OrRd")
plt.legend(loc = 'best')
plt.grid()
plt.title('AVG_GLUCOSE_LEVEL -> AGE')

plt.figure(figsize = (12, 7))
sns.scatterplot(x = data['bmi'], y = data['avg_glucose_level'],  hue = categorical_values['stroke'], palette="OrRd")
plt.legend(loc = 'best')
plt.grid()
plt.title('AVG_GLUCOSE_LEVEL -> BMI')
```
